```python
import csv
import matplotlib.pyplot as plt
import numpy as np

plt.clf()

x1, y1 = np.loadtxt(r'C:\Users\tiff\Documents\PE\250.txt', delimiter='\t', unpack=True)
plt.plot(x1,y1, 'purple', label='249-259 nm')

x2, y2 = np.loadtxt(r'C:\Users\tiff\Documents\PE\436.txt', delimiter='\t', unpack=True)
plt.plot(x2,y2,label='360-370 nm')

x3, y3 = np.loadtxt(r'C:\Users\tiff\Documents\PE\410.txt', delimiter='\t', unpack=True)
plt.plot(x3,y3, 'c', label='410-418 nm')

x4, y4 = np.loadtxt(r'C:\Users\tiff\Documents\PE\365.txt', delimiter='\t', unpack=True)
plt.plot(x4,y4, 'green', label='431-440 nm')

x5, y5 = np.loadtxt(r'C:\Users\tiff\Documents\PE\545.txt', delimiter='\t', unpack=True)
plt.plot(x5,y5, 'orange' ,label='545-555 nm')

x6, y6 = np.loadtxt(r'C:\Users\tiff\Documents\PE\575.txt', delimiter='\t', unpack=True)
plt.plot(x6,y6, 'red' ,label='575-584 nm')

fig = plt.gcf()

fig.set_size_inches(10, 8)


plt.xlabel('Applied Voltage [V]')
plt.ylabel('Retarding Current [microamps]')
plt.title('Retarding Current vs. Voltage')
plt.legend()
plt.show()
```
