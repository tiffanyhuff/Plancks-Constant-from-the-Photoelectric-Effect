```python
import csv
import matplotlib.pyplot as plt
import numpy as np

plt.clf()

x1, y1 = np.loadtxt(r'C:\Users\tiff\Documents\PE\430ND01.txt', delimiter='\t', unpack=True)
plt.plot(x1,y1, label='1')

ry1=np.sqrt(y1)

plt.plot(x1, ry1)

idx = np.argwhere(np.diff(np.sign(y1 - ry1)) != 0)

plt.plot(x1[idx], y1[idx], 'ms', ms=7, label='Nearest data-point method')
print(x1[idx], y1[idx])

#x2, y2 = np.loadtxt(r'C:\Users\tiff\Documents\PE\430ND03.txt', delimiter='\t', unpack=True)
#plt.plot(x2,y2, label='3')

#x3, y3 = np.loadtxt(r'C:\Users\tiff\Documents\PE\430ND07.txt', delimiter='\t', unpack=True)
#plt.plot(x3,y3, label='7')

#x4, y4 = np.loadtxt(r'C:\Users\tiff\Documents\PE\436.txt', delimiter='\t', unpack=True)
#plt.plot(x4,y4, label='436')

plt.xlabel('Applied Voltage [V]')
plt.ylabel('Retarding Current [nA]')
plt.title('Photoelectric Effect')
#plt.annotate('local max', xy=(0.0, 150), xytext=(0.0, 200), c='c')
#ax.set_ylim(-2, 2)
plt.legend()
plt.show()
```
