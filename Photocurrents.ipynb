```python
import csv
import matplotlib.pyplot as plt
import numpy as np

plt.clf()

x1, y1 = np.loadtxt(r'C:\Users\tiff\Documents\PE\250.txt', delimiter='\t', unpack=True)
plt.plot(x1,y1,'mo', ms=5, label='250 nm')


plt.yscale('linear')
plt.grid(True)


plt.xlabel('Applied Voltage [V]')
plt.ylabel('Retarding Current [hA]')
plt.title('Photoelectric Effect')
plt.legend()
plt.show()
```
