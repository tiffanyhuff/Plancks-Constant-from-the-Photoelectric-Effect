```python
import csv
import matplotlib.pyplot as plt
import numpy as np

plt.clf()

x1, y1 = np.loadtxt(r'C:\Users\tiff\Documents\PE\250.txt', delimiter='\t', unpack=True)
plt.plot(x1,y1, label='250')


idx = np.argwhere(np.diff(np.sign(y1 - ry1)) != 0)

plt.plot(x1[idx], y1[idx], 'ms', ms=5, label='Nearest data-point method')

print(x1[idx], y1[idx])


plt.xlabel('Applied Voltage [V]')
plt.ylabel('Retarding Current [nA]')
plt.title('Photoelectric Effect')
plt.legend()
plt.show()
```
